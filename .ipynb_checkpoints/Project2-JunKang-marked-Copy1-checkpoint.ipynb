## Data cleaning: `Deductive imputation` for selected features
---

We will complete the data cleaning using `Deductive Imputation` for the following selected features in both both datasets:
1. Numerical: `Overall Qual`, `Gr Liv Area`, `Garage Area`, `Total Bsmt SF`, `Year Built`, `Full Bath`
2. Categorical: `Foundation`, `BsmtFin Type 1`

```python
df['Garage Area']=df['Garage Area'].fillna(0)
df['Total Bsmt SF']=df['Total Bsmt SF'].fillna(0)
df['BsmtFin Type 1']=df['BsmtFin Type 1'].fillna('None')
test['BsmtFin Type 1']=test['BsmtFin Type 1'].fillna('None')
df.reset_index(inplace=True)
```

It is reasonable to use `Deductive Imputation` and map the null values in `Garage Area`, `Total Bsmt SF` and `BsmtFin Type 1` to '0' or 'None'as these houses are likely to have no garages or basements.

> <font size = 3 color = "crimson"> Addressed all null values? Good to do a check.   </font>

## Feature Engineering: `PolynomialFeatures`
---

We will next use `PolynomialFeatures` to create interaction between the numerical features.

```python
interaction = ['Full Bath','Year Built','Total Bsmt SF','Garage Area','Gr Liv Area','Overall Qual']
```

```python
poly = PolynomialFeatures(degree=2,include_bias=False)
X = poly.fit_transform(df[interaction])
poly.get_feature_names(interaction)
```

## Feature Engineering: `OneHotEncoding`
---

We will convert the categorical columns into a one-hot encoded matrix using the `pd.get_dummies()` method.

```python
dummy = ['Foundation','BsmtFin Type 1']
Z = df[dummy]
Z = pd.get_dummies(Z,columns=dummy,drop_first=True)
Z.head()
```

## Model Prep: Train/test split
---

We will holdout a set of `Train.csv` to test the model using the `train_test_split` function to split our `X` and `y` variables into a training set and a holdout set.

```python
X = pd.DataFrame(X,columns=poly.get_feature_names(interaction))
X = X.join(Z)
y = df['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=50)
```

> <font size = 3 color = "crimson"> Do note that if you further split train into train and validation, and you are imputing by mean/mode/median, then you would need to impute the train and validation set using the  mean/mode/median from the train set.    </font>

## Model Prep: Scaling
---

We will scale our features for `Lasso`, `Ridge` and `ElasticNet`models.

```python
ss = StandardScaler()
ss.fit(X_train)
X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

> <font size = 3 color = "crimson"> ** Baseline for Regression ** The baseline score for regression can be computed by using the average Y value (sale price) as predicted Y. </font>

## Model Prep: Instantiate our models
---

We will be evaluating `LinearRegression`, `LassoCV`, `RidgeCV` and `ElasticNetCV`.

```python
lr = LinearRegression()
lasso = LassoCV(n_alphas=200)
ridge = RidgeCV(alphas=np.linspace(.1, 10000, 100))
enet = ElasticNetCV(alphas=np.arange(0.5, 1.0, 0.005), l1_ratio=0.5, cv=5)
```

## Model Selection: Cross validation
---

We will use `cross_val_score` to evaluate all four models.

```python
lr = LinearRegression()
lasso = LassoCV(n_alphas=200)
ridge = RidgeCV(alphas=np.linspace(.1, 10000, 100))
enet = ElasticNetCV(alphas=np.arange(0.5, 1.0, 0.005), l1_ratio=0.5, cv=5)
```

```python
lr_mean_rmse = np.sqrt(-cross_val_score(lr, X_train, y_train, cv=3, scoring='neg_mean_squared_error').mean())
lr_mean_r2 = cross_val_score(lr, X_train, y_train, cv=3, scoring='r2').mean()
lr_mean_rmse, lr_mean_r2
```

```python
lasso_mean_rmse = np.sqrt(-cross_val_score(lasso, X_train, y_train, cv=3, scoring='neg_mean_squared_error').mean())
lasso_mean_r2 = cross_val_score(lasso, X_train, y_train, cv=3, scoring='r2').mean()
lasso_mean_rmse, lasso_mean_r2
```

> <font size = 3 color = "crimson"> How many features are there in X_train now? There is a convergence warning and you have not obtained the optimal alpha and corresponding scores.  </font>

```python
ridge_mean_rmse = np.sqrt(-cross_val_score(ridge, X_train, y_train, cv=3, scoring='neg_mean_squared_error').mean())
ridge_mean_r2 = cross_val_score(ridge, X_train, y_train, cv=3, scoring='r2').mean()
ridge_mean_rmse, ridge_mean_r2
```

```python
enet_mean_rmse = np.sqrt(-cross_val_score(enet, X_train, y_train, cv=3, scoring='neg_mean_squared_error').mean())
enet_mean_r2 = cross_val_score(enet, X_train, y_train, cv=3, scoring='r2').mean()
enet_mean_rmse, enet_mean_r2
```

## Model Fitting and Evaluation
---

`LassoCV` seems to be doing the best as it has the highest RMSE and R2 score. Hence, we will fit the model to the training data, and evaluate the training and test scores below.

```python
lasso.fit(X_train, y_train)
pred = lasso.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test,pred))
rmse
```

## Data Prep for Prediction
---

We will now prepare the data given in `test.csv` for the model's use.

```python
X = test[interaction]
poly = PolynomialFeatures(degree=2,include_bias=False)
X = poly.fit_transform(X)
Z = test[dummy]
Z = pd.get_dummies(Z,columns=dummy,drop_first=True)
X = pd.DataFrame(X,columns=poly.get_feature_names(interaction))
X = X.join(Z)
ss.fit(X)
X = ss.transform(X)
```

## Prediction
---

We will now fit the data into the the model.

```python
pred = lasso.predict(X)
pred = pd.DataFrame(pred)
result = test['Id']
result = pd.concat([result,pred],axis=1)
result.rename(columns={0:'SalePrice'},inplace=True)
result.to_csv('./output/submission.csv',index=False)
result.head()
```
